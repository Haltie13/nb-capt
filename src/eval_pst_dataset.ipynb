# Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_json("data/pst_scores.json")
df["avg_score"] = df["pseudo_scores"].apply(lambda x: np.mean(x) if len(x) > 0 else 0)
df["group"] = df["audio_path"].apply(lambda x: x.split('/')[-2] if '/' in x else '')
df["speaker"] = df["audio_path"].apply(lambda x: x.split('/')[-1].split('-')[0] if '/' in x else '')
df.head()
```

# Column Information

```python
df.info()
```

# Statistics

```python
df.groupby("is_native")["avg_score"].describe()
```

# Plots

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="avg_score", hue="is_native", kde=True, bins=30, multiple="stack")
plt.title("Distribution of Average Pronunciation Scores")
plt.xlabel("Average Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="is_native", y="avg_score")
plt.title("Comparison of Pronunciation Scores between Native and L2 Speakers")
plt.xlabel("Is Native")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x="group", y="avg_score", hue="is_native")
plt.title("Pronunciation Scores by Group")
plt.xlabel("Group")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()
```
