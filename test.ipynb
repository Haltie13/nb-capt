```python
import torch
from transformers import pipeline, AutoProcessor, AutoModelForCTC, Wav2Vec2CTCTokenizer
import librosa

device="cuda" if torch.cuda.is_available() else "cpu"
```

```python
model_id = "NbAiLab/nb-wav2vec2-300m-bokmaal-v2"
```

```python
from pathlib import Path

ROOT_PATH = Path().resolve().parent
NB300_PATH = ROOT_PATH / "nb-wav2vec2-300m-bokmaal-v2"

NB300_PATH.exists()
```

```python
import json

tokenizer_conig_path = NB300_PATH / "tokenizer_config.json"
with open(tokenizer_conig_path, 'r') as f:
    tokenizer_config = json.load(f)

tokenizer_config
```

```python
tokenizer = Wav2Vec2CTCTokenizer.from_pretrained(
    model_id, 
    **tokenizer_config
)
processor = AutoProcessor.from_pretrained(model_id, tokenizer=tokenizer)
model = AutoModelForCTC.from_pretrained(model_id)
```
